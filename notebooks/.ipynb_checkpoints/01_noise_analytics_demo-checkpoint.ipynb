# Noise Analytics Pipeline  
### Análisis automático de ruido ambiental mediante series temporales

Este notebook presenta un ejemplo de **pipeline de análisis de datos temporales**
aplicado al ruido ambiental.

El objetivo no es tanto el dominio acústico, sino mostrar:
- cómo estructurar un flujo de datos reproducible,
- cómo extraer información relevante de una serie temporal,
- y cómo detectar automáticamente eventos anómalos.

Los datos utilizados son **sintéticos**, diseñados para simular escenarios reales.


## Contexto del proyecto

En muchos estudios de ruido ambiental, el análisis de datos sigue siendo
manual o semi-manual (Excel, inspección visual, umbrales fijos).

Este proyecto propone un enfoque automatizado basado en:
- estadísticas móviles,
- agregaciones temporales,
- y detección adaptativa de eventos.

El pipeline completo está implementado en Python y separado en módulos.


```python
import sys
from pathlib import Path

# Localizamos la raíz del proyecto y añadimos src/ al path
# para poder reutilizar las funciones del pipeline
PROJECT_ROOT = Path("..").resolve()
SRC_PATH = PROJECT_ROOT / "src"

if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))

import pandas as pd
import matplotlib.pyplot as plt

# Importamos únicamente funciones de visualización
# (el procesado se realiza fuera del notebook)
from viz import (
    plot_time_series_with_events,
    plot_heatmap_hour_day,
    plot_event_scatter,
)
```

## Datos de entrada

En este notebook no se regeneran los datos ni se ejecuta el pipeline completo.
Se parte directamente de los resultados ya procesados:

- Serie temporal regularizada a 1 minuto
- Estadísticas móviles calculadas
- Eventos detectados automáticamente

Esto permite centrar la demo en el análisis y la interpretación.


```python
# Cargamos la serie temporal con estadísticas móviles
df = pd.read_csv(
    PROJECT_ROOT / "data/processed/noise_with_rolling.csv",
    parse_dates=["timestamp"]
)

# Cargamos los eventos detectados
events = pd.read_csv(
    PROJECT_ROOT / "data/processed/detected_events.csv",
    parse_dates=["start_time", "end_time"]
)
```

## Serie temporal y detección de eventos

En primer lugar, se visualiza la serie temporal de ruido junto con:
- una media móvil (suavizado),
- y los eventos detectados automáticamente.

Un evento se define como un periodo donde el nivel supera
el comportamiento medio reciente de forma significativa.


```python
plot_time_series_with_events(df, events, site="P1")
```

plot_time_series_with_events(df, events, site="P1")


## Patrones temporales: hora del día vs día

El siguiente mapa de calor permite identificar patrones globales:
- diferencias entre día y noche,
- días más conflictivos,
- regularidades temporales.

Es una forma compacta de resumir grandes volúmenes de datos.


```python
plot_heatmap_hour_day(df, site="P1")
```

## Caracterización de eventos

Por último, se analiza la relación entre:
- la duración de los eventos,
- y su nivel máximo.

Esto permite diferenciar entre eventos cortos e intensos
y eventos más largos pero menos extremos.


```python
plot_event_scatter(events)
```

## Conclusiones

- El pipeline permite analizar series temporales de ruido de forma automatizada.
- La detección de eventos se adapta al contexto temporal sin usar umbrales fijos.
- Las visualizaciones facilitan la interpretación y la toma de decisiones.
- El enfoque es extensible a otros tipos de señales temporales.
